BUSQUEDA EN ANCHURA

```python
#Definición del problema
#Espacio_estados = A, B, C, D, E, F, G, H, I, J
#Estado_inicial = A
#Estado_objetivo = H
#Acciones = M_Iz, M_Cen, M_Der
```

```python
#Funciones sucesoras [Accion,Estado,Costo]
A= [["M_iz","B",1],["M_Cen","C",1],["M_der","D",1]]
B= [["M_iz","E",1],["M_der","F",1]]
C= [["M_iz","G",1],["M_der","H",1]]
D= [["M_iz","I",1],["M_der","J",1]]

#Función objetivo
#Estado actual == "H" => Objetivo conseguido
```

Pocisiones:

```python
A[1]
```

```python
A[0][0]
```

Funcion buscar: recorrido en anchura (primer nivel es ["M_iz","B",1],["M_Cen","C",1],["M_der","D",1])


```python
def buscar():
  for i in A: 
    if i[1]=="H":
      print("Solución encontrada en ",i[1])
    else:
      print("Solución NO encontrada en ",i[1])
  for j in B:
    if j[1]=="H":
      print("Solución encontrada en ",j[1])
    else:
      print("Solución NO encontrada en ",j[1])
  for k in C:
    if k[1]=="H":
      print("Solución encontrada en ",k[1])
    else:
      print("Solución NO encontrada en ",k[1])
buscar()
```

Si se hace el for interno es por PROFUNDIDAD y si se hace el for externo es ANCHURA

Cuando se encuentra la solución la idea es romper el ciclo con un brake


La solución es A, C, H y de ahi se saca el costo del camino
